### **DATA VISUALIZATION  and  INSIGHTS**

Visualizations to better understand the data




```python
# Bar plot of the count of customers by gender:

sns.countplot(x='sex', data=data)
plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Inference: 

The count of male customers is slightly higher than the count of female customers.

```python
# Distribution plot of tip

sns.distplot(data['tip'], kde=False)
plt.title('Distribution of Tips')
plt.show()
```

inference:

tips tend to be around 2-4 dollar, with some larger tips up to $10.

```python
# Violin plot of total bill

sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Total Bill Distribution by Day')
plt.show()
```

INFERENCE:

The total bill tends to be higher on weekends (Fri, Sat, Sun) than on weekdays (Thur).

```python
# Count plot of customers who smoke by day of the week:

sns.countplot(x='day', data=data, hue='smoker')
plt.title('Count of Customers who Smoke by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
```

INFERENCE:

A majority of customers who smoke tend to visit the restaurant on Fridays and Saturdays. 

There are also some non-smokers who visit the restaurant on those days.

```python
# Joint plot of total bill and tip

sns.jointplot(x='total_bill', y='tip', data=data)
plt.title('Relationship between Total Bill and Tip')
plt.show()
```

INFERENCE:

There is a positive relationship between total bill and tip, which makes sense since people may **tip more for larger bills.**

```python
 # boxplot of tip amount for smokers and non-smokers:

sns.boxplot(x="smoker", y="tip", data=data)
plt.title("Tip amount for smokers and non-smokers")
plt.show()
```

INFERENCE:



*   **smokers tend to leave slightly higher tips than non-smokers**, with a median tip amount of around 3 dollars for smokers and around 2.75 dollars for non-smokers.




```python
#  scatterplot between time and tip using a scatter plot:

sns.scatterplot(data=data, x="time", y="tip")
```

INFERENCE:

There is no significant difference between the tips given during lunch and dinner. 

However, we can observe that there are some **high tip values during dinner time**, indicating that dinner customers tend to be more generous with tipping than lunch customers.

```python
# plot of average tip amount by day:

# Calculate the average tip by day
avg_tip_by_day = data.groupby('day')['tip'].mean().reset_index()

# Create a barplot to visualize the average tip by day
sns.barplot(x='day', y='tip', data=avg_tip_by_day)

```

INFERENCE:

The average **tip amount is highest on Sundays and lowest on Thursdays. **

This could suggest that customers tend to be more generous with tips on weekends, possibly because they are more relaxed and have more time to enjoy their meals.

```python
# plot showing the relationship between sex and tip:

sns.boxplot(x="sex", y="tip", data=data)
plt.title("Boxplot of Tip Amount by Gender")
plt.show()
```

INFERENCE:

On average, **males tend to leave slightly higher tips than females. **

However, there is a lot of overlap in the distribution of tip amounts between genders, and there are also some outliers in both groups. 

Therefore, while there may be a slight gender difference in tipping behavior, it is not a strong or definitive pattern.

```python
# plot showing the relationship between size and tip:

plt.scatter(data['size'], data['tip'])
plt.xlabel('Size')
plt.ylabel('Tip')
plt.title('Relationship between Size and Tip')
plt.show()
```

INFERENCE:

There is a positive correlation between the size of the group and the tip amount. 

**As the size of the group increases, the tip amount tends to increase** as well.

This could be because larger groups tend to order more food and drinks, resulting in a higher bill and a higher tip.

```python
# box plot that shows the distribution on total bill, gender and day

sns.set_style('whitegrid')
sns.catplot(x='day', y='total_bill', hue='sex', kind='box', data=data, height=4, aspect=1.5)
plt.title('Total Bill by Day and Gender')
plt.show()
```

INFERENCE: 



*  On average, males tend to have a slightly higher total bill than females on all days except for Friday, where the difference is negligible.
*  total bill is generally higher on weekends (Saturday and Sunday) compared to weekdays (Thursday and Friday), with Saturday having the highest median total bill.


Overall, this plot suggests that the combination of gender and day can have an impact on the total bill, which can be useful information for businesses in the hospitality industry.

```python
# 2D density plot with contours that shows the distribution of tips depending on total bill 

sns.kdeplot(x='total_bill', y='tip', data=data, shade=True, thresh=0.05, cmap="viridis")
sns.jointplot(x='total_bill', y='tip', data=data, kind='kde', shade_lowest=False, cmap="viridis")
plt.show()
```

INFERENCE:



*   The 2D density plot shows the distribution of total bill and tip across different values, with darker areas indicating higher density.
*   The contour lines in the plot indicate the areas of highest density.




*   There is a strong positive correlation between total bill and tip, with higher tips being given on higher total bills.
*   The joint plot with kde representation also shows the bivariate distribution of total bill and tip, with darker areas indicating higher density.


*   The distribution is skewed towards the lower end of the total bill and tip values, indicating that most of the tips are given on lower bills.







```python
# kernel density estimate plot that shows the distribution of tips depending on total bill and time

import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=data, x="total_bill", hue="time", fill=True)
plt.title("Kernel Density Estimate Plot of Total Bill by Time")
plt.show()

#The resulting plot shows the estimated probability density function of the total bill for each time category. 
```

INFERENCE:

The distribution of the total bill is similar for both lunch and dinner, with the peak around $15-$20. 

However, there are more high total bill values for dinner than for lunch.

```python
# swarm plot (categorical plot) depending on total bill, gender, smoker, and day:

sns.catplot(x='day', y='total_bill', hue='sex', col='smoker', data=data, kind='swarm')
```

INFERENCE:

On Friday and Saturday nights, there are more male customers who smoke and pay higher bills compared to other days. 

Also, on Sundays, non-smoking females pay higher bills than smoking females.

```python
# the distribution of tips and total bill by gender and day

sns.set(style="ticks")

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a FacetGrid to plot multiple distributions on one plot
g = sns.FacetGrid(tips, row="day", col="sex", hue="sex", margin_titles=True)

# Plot the distributions
g.map(sns.kdeplot, "total_bill", "tip", alpha=.7)
g.add_legend()

# Set titles for each subplot
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Tips and Total Bill by Gender and Day', fontsize=16)

# Show the plot
plt.show()
```

INFERENCE:



*   There is a positive correlation between total bill and tip

*   The distribution of tips and total bill is higher for males than females, and the highest on Saturdays. 


*   Thursdays have the lowest tips and total bills.





```python
# Distribution plot showing the distribution of tips and total bill by smoker and day 

sns.displot(data=data, x="tip", col="smoker", row="day", hue="sex", kind="kde")

```

INFERENCE:

*   On Fridays and Saturdays, the distribution of tips is wider and skewed towards higher values, whereas on other days, the distribution is more symmetric. 

*   non-smokers tend to give slightly higher tips than smokers,on weekends and the difference is more pronounced for females.






```python
# Visual Analysis dePending on tips, total bill ,gender, smoker and day 


# scatterplot of tips vs total bill, with hue for gender and style for smoker
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker")
plt.title("Scatterplot of Tips vs Total Bill by Gender and Smoker Status")
plt.show()

# barplot of average tips by day, with hue for gender and smoker status
sns.barplot(x="day", y="tip", data=tips, hue="sex", hue_order=["Male", "Female"], estimator=np.mean, ci=None)
plt.title("Average Tips by Day and Gender")
plt.show()

# boxplot of total bill by smoker status and day, with hue for gender
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", hue_order=["Male", "Female"], palette="Set2", 
            notch=True, showcaps=True, showfliers=True, showbox=True)
plt.title("Boxplot of Total Bill by Smoker Status and Day, with Gender")
plt.show()

# violinplot of tips by smoker status and day, with hue for gender
sns.violinplot(x="day", y="tip", data=tips, hue="sex", hue_order=["Male", "Female"], split=True, inner="stick")
plt.title("Violinplot of Tips by Smoker Status and Day, with Gender")
plt.show()
```

INFERENCE:



1.   As the total bill increases, so does the tip amount. However, there are some outliers with a high total bill and low tip, as well as some with a low total bill and high tip. We can also observe that there are more male customers than female customers, and that there are more non-smokers than smokers.

2.   The average tip amount is highest on Sundays and lowest on Saturdays. We can also observe that there is no significant difference in average tip amount between male and female customers, but that smokers tend to leave slightly higher tips than non-smokers.

3.   The median total bill is generally higher for non-smokers than smokers, and that the median total bill is highest on Saturdays and lowest on Fridays. We can also observe that male customers tend to have a higher total bill than female customers, with the exception of Sundays.

4.   The distribution of tip amounts is wider for non-smokers than smokers, and that the highest tip amounts are generally left on Sundays. We can also observe that there is no significant difference in tip amounts between male and female customers.

*   There is a **moderate positive correlation between total_bill and size**, indicating that larger groups tend to have higher bills.

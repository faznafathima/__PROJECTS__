```python

#sending request to get the web page and to get the data in html format, using soup and passing html parser as a parameter to get the data in that format

#beautiful soup is used for the scraping
page=requests.get(web_url3,headers=headers)
soup=BeautifulSoup(page.content,'html5lib')
#print(soup)
```

**Web Crawling**

```python
page_links=[]
links=soup.find_all('a',{'class':'a-link-normal s-no-outline'})
for l in links:
  page_links.append("https://www.amazon.in/"+l.get('href'))
print(page_links)
print()
print('number of items within the page : ',len(page_links))
```

```python
url = "https://www.amazon.in/s?k=laptop"
amz_cont = requests.get(url,headers=headers)
soup = BeautifulSoup(amz_cont.content, 'html5lib')
soup
```

**Web Scraping**

```python
amazon_laptops=[]
for i in page_links:
  page_link=i
  item_page=requests.get(page_link,headers=headers)
  soup=BeautifulSoup(item_page.content,'html.parser')

  try:
    Item_name=soup.find(id='productTitle').get_text()
  except AttributeError:
    continue

  
  #Price=soup.find(id='tp_price_block_total_price_ww')
  #pricee=soup.find(id='tp_price_block_total_price_ww').get_text()
  #Price=float(pricee[1:3]+pricee[4:7])
  #Price=pricee[1:3]+pricee[4:7]

  #pricee = soup.find('span','a-price')
  #Price = pricee.find('span','a-offscreen').text

  try:
    #price
    pricee = soup.find('span','a-price')
    Price = pricee.find('span','a-offscreen').text
  except AttributeError:
    continue

  
  Image=soup.find(id='landingImage').attrs['src']

  #Rating=soup.find(id='acrPopover').get_text()
  #Rating=soup.find(i.text)
  Rating = soup.find("span", {"class": "a-icon-alt"})
  Rating=Rating.text

  About_item=soup.find(id='feature-bullets').get_text()

  amazon_laptops.append([Item_name.strip(),Price,Image,Rating,About_item.strip()])

print(amazon_laptops)




```

**Creating the DataFrame**

```python
df=pd.DataFrame(amazon_laptops,columns=['Item_name','Price','Image','Rating','About_item'])
df
```

**INSERTING DATA INTO SQLITE-3**

```python
for i  in df.itertuples():
  print(i)
```

```python
#removing the index
for j in df.itertuples():
  print(j[1:])
```

```python
conn = sqlite3.connect(":memory:") # database(stored in variable conn) is created temprorarily in memory
cursor = conn.cursor()
cursor.execute("""
    CREATE TABLE df(Item_name, Price, Image, Rating, About_item)
""")
for row in df.itertuples():
    insert_table = """
    INSERT INTO df(Item_name, Price, Image, Rating, About_item) \
    VALUES (?,?,?,?,?)
"""
    cursor.execute(insert_table, row[1:])
#conn.close()
conn.commit()
```

**Retrieving data from TABLE df and displaying the first two item records using SELECT command-Data Query Language**

```python
retrieving = conn.execute("SELECT Item_name, Price, Image, Rating, About_item FROM df")
for row in retrieving:
   print("Item_name = ", row[0],"\n")
   print("Price = ", row[1],"\n")
   print("About_item = ", row[4], "\n")
   print()
```

# **MULTICLASS SVM ON IRIS DATASET**

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(r'/content/IRIS.csv')
data
```

```python
pd.get_dummies(data,columns=['species'])      
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data.drop('species',axis=1))
scaled_features = scaler.transform(data.drop('species',axis=1))
data_scaled = pd.DataFrame(scaled_features,columns=data.columns[:-1])
data_scaled.head(10)
```

```python
X=data_scaled
y=data['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,mean_absolute_error,mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)
```

```python
y_test.value_counts()
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
